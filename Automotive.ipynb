```python
df['Number of Doors'] = df['Number of Doors'].fillna(
    df['Number of Doors'].mean()
)
```

```python
df.drop_duplicates(keep='first', inplace=True)

```

```python
valid_trans = ['AUTOMATIC', 'MANUAL', 'AUTOMATED_MANUAL', 'DIRECT_DRIVE']
```

```python
df= df[df['Transmission Type'].isin(valid_trans)]
```

```python
df = df[df['Engine Cylinders'] >= 1].copy() 
```

### Are there outliers present in key numerical features?

```python
outliers_dict = {}

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    outliers_dict[col] = outliers.shape[0]  # number of outliers

# Show number of outliers per column
outliers_summary = pd.DataFrame.from_dict(outliers_dict, orient='index', columns=['Num_Outliers'])
print(outliers_summary)
```

```python
plt.figure(figsize=(10, 6))
outliers_summary.sort_values(by='Num_Outliers', ascending=False).plot(
    kind='bar',
    legend=False,
    color='tomato',
    edgecolor='black'
)

plt.title('Number of Outliers per Numeric Column')
plt.ylabel('Number of Outliers')
plt.xlabel('Numeric Columns')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

detected the outliers by the IQR method by iterating over the numeric columns abd detecting each column and then visulaized the number of outliers in each column using a bar chart
